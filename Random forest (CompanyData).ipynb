# Data Processing

```python
rf['ShelveLoc'].unique(),rf['Urban'].unique(),rf['US'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
```

```python
rf['ShelveLoc']=LE.fit_transform(rf['ShelveLoc'])
rf['Urban']=LE.fit_transform(rf['Urban'])
rf['US']=LE.fit_transform(rf['US'])
```

```python
rf.dtypes
```

```python
rf
```

# converting sales into categorical variable

```python
rf.insert(11,"sales_variable",'')
rf
```

```python
for i in range(0,len(rf['Sales'])):
    if rf['Sales'][i]>=11.0:
        rf['sales_variable'][i] = 'High'
    elif rf['Sales'][i]<=6.0:
        rf['sales_variable'][i] = 'Low'
    else:
        rf['sales_variable'][i] = 'Medium'
```

```python
rf
```

# Model building

```python
X=rf.iloc[:,1:-1]
Y=rf.iloc[:,-1]
```

```python
X.head()
```

```python
Y.head()
```

# split into training and testing data

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=7)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# Grid Search CV To find Best Hyperparameter

```python
from sklearn.model_selection import GridSearchCV
```

```python
mod=RandomForestClassifier()
parameters={
    
    "n_estimators":[130, 140, 150, 160],
    "max_features":[4,5,6,7]
}
```

```python
Grid_DT=GridSearchCV(mod,parameters,cv=10)
```

```python
Grid_DT.fit(x_train,y_train)
```

```python
Grid_DT.best_score_
```

```python
Grid_DT.best_params_
```

```python
final_mod=RandomForestClassifier(max_features=7, n_estimators=130)
```

```python
final_mod.fit(x_train,y_train)
```

```python
print(classification_report(y_test,final_mod.predict(x_test)))
```
